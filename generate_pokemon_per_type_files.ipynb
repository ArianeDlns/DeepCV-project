```python
import pandas as pd
import shutil
import os
import numpy as np
from PIL import Image, ImageOps, ImageCms

import glob
```

```python
df_poke = pd.read_csv("data/pokemon.csv")

df_poke
```

```python
pokemon_name_list = df_poke["Name"].tolist()
pokemon_type_list = df_poke["Type1"].unique()
```

```python
type_to_pokemon_list = {}

for pokemon_type in pokemon_type_list :
    type_to_pokemon_list[pokemon_type] = []

for _, row in df_poke.iterrows():
    pokemon_name = row[0]
    pokemon_type1 = row[1]
    pokemon_type2 = row[2]
    
    type_to_pokemon_list[pokemon_type1].append(pokemon_name)
    if pokemon_type2 in pokemon_type_list: #A Pokemon might only have one type
        type_to_pokemon_list[pokemon_type2].append(pokemon_name)

type_to_pokemon_list
```

```python
IMAGES_DIR = "./data/images/images/*.*"

for image_path in glob.glob(IMAGES_DIR):
    pokemon_name = image_path.split('\\')[-1].split('.')[0]
    print(pokemon_name)
    
    # Get pokemon type from dataframe
    pokemon_type = df_poke.loc[df_poke['Name'] == pokemon_name].iloc[0]['Type1']
    
    new_path = 'data/pokemon_png/' + pokemon_name + '.' + 'png'
    print('New path: ', new_path)
    
    shutil.copyfile(image_path, new_path)
```

```python
#converts all files in pokemon_png to no transparent images (remove transparancy)

PNG_DIR = "./data/pokemon_png/*.*"

for filename in glob.glob(PNG_DIR):

        print(filename)
        
        png = Image.open(filename).convert('RGBA') #Normally all png have a transparent background but some are in Palette mode so need to convert
        png.load()

        background = Image.new("RGB", png.size, (255, 255, 255))
        background.paste(png, mask=png.split()[3])        

        os.remove(filename)

        background.save(filename[:-4] + ".png")

        
```

```python
#Create train val dataset for CNN training

os.makedirs("data/pokemon_per_type")

os.makedirs("data/pokemon_per_type/train")

os.makedirs("data/pokemon_per_type/val")

for pokemon_type in pokemon_type_list:

    pokemon_type_folder_paths = []

    pokemon_type_folder_paths.append("data/pokemon_per_type/train/" + pokemon_type)
    os.makedirs(pokemon_type_folder_paths[-1])

    pokemon_type_folder_paths.append("data/pokemon_per_type/val/" + pokemon_type)
    os.makedirs(pokemon_type_folder_paths[-1])

    for pokemon_name in type_to_pokemon_list[pokemon_type]:

        path = pokemon_type_folder_paths[np.random.binomial(1,0.2)]
        shutil.copy("data/pokemon_png/" + pokemon_name + ".png", path)
```
